```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

# Das linalg-Paket
Numpy bietet vielfältige Funktionen für die lineare Algebra

```python
M = np.random.randint(low=2, high=19, size=(3,3))
M
```

## Eigenwert
Den Faktor, um wie viel der Vektor nach Multiplikation mir der Matrix länger oder kürzer geworden ist, nennt man Eigenwert.

```python
lambda_ = np.linalg.eigvals(M)
lambda_
```

## Eigenvektor
Ein Eigenvektor einer Abbildung ist in der linearen Algebra ein vom Nullvektor verschiedener Vektor, dessen Richtung durch die Abbildung nicht verändert wird. Ein Eigenvektor wird also nur skaliert und man bezeichnet den Skalierungsfaktor als Eigenwert der Abbildung.

```python
np.linalg.eig(M)
```

## Matrix-Norm
Matrix or vector norm.

```python
np.linalg.norm(M)
```

## Vektor-Transformation

```python
import math

def print_quiver(ax, origin, vector, label):
    ax.quiver(*origin, *vector, color=['g', 'b', 'r'], linewidth=1, angles='xy', scale_units='xy', scale=1, label=label)
    ax.axis([-2, 2, 0, 2])
    ax.legend()


def rotate(degrees, v):
    theta = math.radians(degrees)
    _cos, _sin = np.cos(theta), np.sin(theta)
    j = np.array([[_cos, _sin], [-_sin, _cos]])
    return v @ j


def reflect(v: np.ndarray) -> np.ndarray:
    return v @ np.array([[-1, 0], [0, 1]])


def scale(v: np.ndarray, value: float) -> np.ndarray:
    return v @ np.array([[value, 0], [0, value]])


fig, (ax, ax2, ax3, ax4) = plt.subplots(4, figsize=(20, 10))
origin = [0, 0]
v1 = np.array([2, 1])
print_quiver(ax, origin, v1, label="Original")


v2 = scale(v1, 0.5)
v3 = rotate(90, v1)
v4 = reflect(v1)

print_quiver(ax3, origin, v3, label="Rotated 90°")
print_quiver(ax4, origin, v4, label="Reflected")
print_quiver(ax2, origin, v2, label=f"Scaled {v2}")
plt.legend()

plt.show()
```
