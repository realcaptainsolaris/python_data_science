# Integration von Vektoren

Die Integration von Vektoren bedeutet, dass man die einzelnen Komponenten eines Vektors integriert. Das Prinzip ist √§hnlich wie bei der Integration von Zahlenfunktionen, nur dass man es hier f√ºr jede Komponente des Vektors einzeln durchf√ºhrt.

### **1. Ein Vektor besteht aus Komponenten**

Ein Vektor \(\vec{F}(t)\) kann in einem 2D- oder 3D-Raum dargestellt werden und besteht aus mehreren **Komponenten**. Zum Beispiel:

$
\vec{F}(t) = \begin{bmatrix} F_1(t) \\ F_2(t) \end{bmatrix} \quad \text{(im 2D-Raum)} \quad \text{oder} \quad \vec{F}(t) = \begin{bmatrix} F_1(t) \\ F_2(t) \\ F_3(t) \end{bmatrix} \quad \text{(im 3D-Raum)}.
$

Jede Komponente $F_1(t), F_2(t), F_3(t)$ ist eine Funktion von \(t\).


### **2. Die Integration wird komponentenweise durchgef√ºhrt**

Um den gesamten Vektor zu integrieren, integriert man jede **Komponente** einzeln:

$
\int \vec{F}(t) \, dt = \begin{bmatrix} \int F_1(t) \, dt \\ \int F_2(t) \, dt \\ \int F_3(t) \, dt \end{bmatrix}.
$

---

### **3. Beispiel**

Betrachten wir einen Vektor im 2D-Raum:

$
\vec{F}(t) = \begin{bmatrix} t \\ t^2 \end{bmatrix}.
$

Um diesen Vektor zu integrieren, berechnen wir die Integrale der beiden Komponenten separat:

1. **Erste Komponente**: $F_1(t) = t$
   $
   \int t \, dt = \frac{t^2}{2} + C_1
   $

2. **Zweite Komponente**: $F_2(t) = t^2$
   $
   \int t^2 \, dt = \frac{t^3}{3} + C_2
   $

Der integrierte Vektor ist dann:

$
\int \vec{F}(t) \, dt = \begin{bmatrix} \frac{t^2}{2} + C_1 \\ \frac{t^3}{3} + C_2 \end{bmatrix}.
$


### **4. Konstanten der Integration**

- Bei der Integration erh√§lt man **Integrationskonstanten** $(C_1, C_2, \dots $), weil die Ableitung einer Konstante $0$ ist.  
- Jede Komponente hat ihre eigene Konstante, weil sie unabh√§ngig voneinander integriert wird.


### **Zusammenfassung**
- Bei der Integration von Vektoren integriert man **jede Komponente separat**.
- Das Ergebnis ist ein Vektor, dessen Komponenten die Integrale der urspr√ºnglichen Funktionen sind.
- Die Konstanten der Integration m√ºssen f√ºr jede Komponente separat ber√ºcksichtigt werden.


### **2D-Beispiel**

Gegeben ist ein **2D-Vektor** $\vec{v}(t)$ mit den Komponenten:
$
\vec{v}(t) = \begin{bmatrix} v_x(t) \\ v_y(t) \end{bmatrix}
$
wobei:

- $ v_x(t) = t $ (die x-Komponente)  
- $ v_y(t) = t^2 $ (die y-Komponente)  

Integriere diesen Vektor √ºber das Zeitintervall $ t \in [0, 2] $ und stelle den resultierenden **integrierten Vektor** graphisch im 2D-Raum dar.

$
\vec{v}_{\text{int}} = \begin{bmatrix} 2.0 \\ 2.67 \end{bmatrix}.
$

- **x-Komponente**: $ \int_0^2 t \, dt = 2.0 $ 
- **y-Komponente**: $ \int_0^2 t^2 \, dt = 2.67 $  

Der gr√ºne Pfeil zeigt den **integrierten Vektor** im 2D-Raum. üòä

```python
import numpy as np
import matplotlib.pyplot as plt


# 2D-Vektorfunktionen definieren
def v_x_2d(t):
    return t  # x-Komponente

def v_y_2d(t):
    return t**2  # y-Komponente

# Integrationsgrenzen
t_start = 0
t_end = 2

# Integration jeder Komponente
# The quad function performs numerical integration for single-variable 
# functions over a specified range [a,b][a,b].
x_integral_2d, _ = quad(v_x_2d, t_start, t_end)
y_integral_2d, _ = quad(v_y_2d, t_start, t_end)

# Integrierter Vektor
vector_integral_2d = np.array([x_integral_2d, y_integral_2d])

# Ausgangsvektor
x_component = v_x_2d(t_end)
y_component = v_y_2d(t_end)
vector_original = np.array([x_component, y_component])

# Plot des Vektors im 2D-Raum
fig, ax = plt.subplots()
ax.quiver(0, 0, x_integral_2d, y_integral_2d, angles='xy', scale_units='xy', scale=1, color='g', label='Integrierter Vektor')
ax.quiver(0, 0, vector_original[0], vector_original[1], angles='xy', scale_units='xy', scale=1, color='b', label='Ausgangsvektor')

# Achseneinstellungen
ax.set_xlim(-4, 4)
ax.set_ylim(-1, 4)
ax.set_aspect('equal')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.xlabel('X-Achse')
plt.ylabel('Y-Achse')
plt.title('Integrierter Vektor in 2D')
plt.legend()

plt.show()

vector_integral_2d
```
