# Fast Fourier Transformation

By applying the FFT and inverse FFT, we successfully remove noise from a simulated audio signal. 
This example demonstrates how the FFT is a practical tool for real-world signal processing challenges, such as noise reduction, frequency analysis, and feature extraction. The approach can be extended to work with actual audio files or other types of time-series data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generating a Sample Signal with Noise

Simulate a time-domain signal consisting of two sine waves, representing the desired audio frequencies, and introduce random noise to mimic real-world conditions.

```python
# Sampling parameters
sampling_rate = 1000  # Hz
duration = 2  # seconds
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Pure signal (two sine waves at 50 Hz and 120 Hz)
signal = np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 120 * t)

# Add random noise
noise = 2 * np.random.normal(size=t.shape)
noisy_signal = signal + noise

# Plot the noisy signal
plt.figure(figsize=(10, 4))
plt.plot(t, noisy_signal, label="Noisy Signal")
plt.title("Time Domain: Noisy Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

## Applying FFT to Transform to the Frequency Domain

To analyze the frequency content of the signal, we apply the FFT. This function transforms the signal into its frequency components, and numpy.fft.fftfreq generates the corresponding frequency values. The resulting spectrum is plotted to display the signal's amplitude for each frequency.

```python
# Apply FFT
fft_values = np.fft.fft(noisy_signal)
frequencies = np.fft.fftfreq(len(fft_values), d=1/sampling_rate)

# Plot the frequency spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_values)[:len(frequencies)//2])
plt.title("Frequency Domain: Noisy Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```

## Identifying and Filtering Noise

To remove noise, we create a filter by isolating the desired frequency ranges (40–60 Hz and 110–130 Hz). A Boolean mask is constructed to retain only the FFT values within these ranges. All other frequency components are set to zero. The filtered frequency spectrum is then plotted to confirm that noise has been removed.

```python
# Create a frequency mask to filter out noise
frequency_mask = (frequencies > 40) & (frequencies < 60) | (frequencies > 110) & (frequencies < 130)
filtered_fft_values = np.where(frequency_mask, fft_values, 0)

# Plot the filtered frequency spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(filtered_fft_values)[:len(frequencies)//2])
plt.title("Frequency Domain: Filtered Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```

## Applying Inverse FFT to Reconstruct the Signal

inally, the filtered signal is transformed back to the time domain using the inverse FFT. This operation reconstructs the cleaned signal by combining only the retained frequency components

```python
# Apply inverse FFT
filtered_signal = np.fft.ifft(filtered_fft_values).real

# Plot the cleaned signal
plt.figure(figsize=(10, 4))
plt.plot(t, filtered_signal, label="Filtered Signal")
plt.title("Time Domain: Cleaned Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```
