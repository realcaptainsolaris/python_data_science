## Verschiedene Verteilungen mit `np.random` DOZENT
In diesem Notebook werden verschiedene Verteilungen mit NumPy erzeugt und visualisiert. Die Visualisierungen werden mit Matplotlib erstellt.

```python
import numpy as np
import matplotlib.pyplot as plt

# Zum besseren Stil
# plt.style.use('seaborn')
```

### 1. Gleichverteilung (`np.random.uniform`)
Erzeugt Werte gleichmäßig verteilt in einem bestimmten Bereich.

```python
# Gleichverteilung
data_uniform = np.random.uniform(low=0.0, high=10.0, size=1000)

# Visualisierung
plt.hist(data_uniform, bins=20, color='skyblue', edgecolor='black')
plt.title('Gleichverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.show()
```

```python
M = np.random.uniform(low=0, high=1000, size=(200, 2))

plt.hist(M, bins=20)
plt.title('Gleichverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.show()
```

### 2. Normalverteilung (`np.random.normal`)
Erzeugt Werte aus einer Normalverteilung mit einem Mittelwert und einer Standardabweichung.

```python
# Normalverteilung
data_normal = np.random.normal(loc=0.0, scale=1.0, size=1000)  # auch hier geht z.b. size=(399, 34)

# Visualisierung
plt.hist(data_normal, bins=20)
plt.title('Normalverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.show()
```

### 3. Binomialverteilung (`np.random.binomial`)
Erzeugt Werte aus einer Binomialverteilung, die die Anzahl der Erfolge in einer Reihe von Experimenten beschreibt.

```python
# Binomialverteilung

# Dozenteninfo
# Interpretation eines Balkens: der Balken für k=5 (5 Erfolge) hat eine Höhe von 250:
# in 250 von 1000 Simulationen gab es genau 5 Erfolge in 10 Versuchen.
# Da p=0.5 ist, ist dies der wahrscheinlichste Wert.

# Beispiel: 
# Produkttest in der Qualitätskontrolle: Eine Fabrik produziert Glühbirnen mit einer Fehlerquote von 50% (p = 0.5), 
# und ein Ingenieur testet 10 zufällig ausgewählte Glühbirnen.
# nimmt 10 Glühbirnen pro Test, aber wiederholt den Test 1000 Mal.

n, p = 10, 0.5  # Anzahl der Versuche, Erfolgswahrscheinlichkeit zwischen 0 und 1

data_binomial = np.random.binomial(n, p, size=1000)

# Visualisierung
plt.hist(data_binomial, bins=range(n+2), align='left', color='salmon', edgecolor='black')
plt.title('Binomialverteilung')
plt.xlabel('Anzahl Erfolge')
plt.ylabel('Häufigkeit')
plt.show()
```

### 4. Poissonverteilung (`np.random.poisson`)
Erzeugt Werte aus einer Poissonverteilung, die seltene Ereignisse modelliert.

```python
# Poissonverteilung

# Beispiel:
# Ein Callcenter erhält im Durchschnitt 4 Anrufe pro Minute – die Poissonverteilung modelliert, 
# wie viele Anrufe tatsächlich in 1000 zufälligen Minuten eingehen, wobei es manchmal 
# weniger oder mehr als 4 sein können.

lambda_param = 4  # Erwartungswert der Verteilung, 4 Ereignisse pro Einheit (z. B. 4 Anrufe pro Minute).
data_poisson = np.random.poisson(lambda_param, size=1000) # Erzeugt 1000 Zufallswerte, die der Poissonverteilung mit λ=4 folgen.

# Visualisierung
plt.hist(data_poisson, bins=range(0, max(data_poisson)+2), align='left', color='gold', edgecolor='black')
plt.title('Poissonverteilung')
plt.xlabel('Wert') # Anzahl der Ereignisse (z. B. wie viele Anrufe in einer Minute).
plt.ylabel('Häufigkeit') # Wie oft jede Anzahl von Ereignissen in den 1000 Simulationen vorkam.
plt.show()
```

### 5. Exponentialverteilung (`np.random.exponential`)
Erzeugt Werte aus einer Exponentialverteilung, die die Zeit zwischen zufälligen Ereignissen beschreibt.

```python
# Exponentialverteilung
scale = 1.0  # 1/Lambda
data_exponential = np.random.exponential(scale, size=1000)

# Visualisierung
plt.hist(data_exponential, bins=20, color='purple', edgecolor='black')
plt.title('Exponentialverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.show()
```

# Aufgabe: Simulation und Analyse einer Normalverteilung mit NumPy

Ein Unternehmen möchte die Produktionszeit einer Maschine analysieren. Es wird angenommen, dass die Produktionszeit pro Produkt normalverteilt ist. 

1) Erzeuge mit np.random.normal eine Stichprobe von 1000 Produktionszeiten, wobei:

    Der Mittelwert der Produktionszeit mü bei 50 Minuten liegt.
    Die Standardabweichung sigma 10 Minuten beträgt.

2) Plotte die Häufigkeitsverteilung

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# 1. Simulation der Produktionszeiten
mu, sigma = 50, 10  # Mittelwert und Standardabweichung
data = np.random.normal(mu, sigma, 1000)

# 2. Visualisierung
# Histogramm
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Normalverteilungskurve (PDF)
x = np.linspace(mu - 6*sigma, mu + 6*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, color='red', lw=2, label='Theoretische Normalverteilung')

plt.title('Produktionszeiten')
plt.xlabel('Produktionszeit (Minuten)')
plt.ylabel('Häufigkeit')
plt.show()
```
