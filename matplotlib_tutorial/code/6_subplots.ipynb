# Subplots

Subplots bedeuten eine Gruppe kleinerer Axes (wobei jede Axes ein Plot ist), die zusammen in einer einzigen Figur existieren können. Man kann damit zum Beispiel zwei oder mehrere Plots nebeneinander ausführen. Wir können das in Matplotlib mit der Methode `subplots` erreichen.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(loc=3, scale=3, size=(10,))
y = np.random.normal(loc=3, scale=3, size=(10,))
x, y
```

## Subplots erzeugen
der Rückgabewert von `subplots()` ist ein Tupel bestehend aus der `Figure` und einem `numpy.ndArray` mit den geschaffenen `Axes-Objekte`. 

An die Funktion selbst übergeben wir im ersten Parameter die Anzahl der Zeilen, im zweiten die Anzahl der Spalten. Das heisst, die Subplots entstehen immer als Grid. 

In die `Axes-Objekte` kann man dann plotten, in dem man auf den Index des `ax`-Arrays zugreift.

## Subplots beschriften
wir können jedem Subplot eigene Achsenbeschriftungen und einen Titel geben. Dazu gibt es die Methoden `set_title`, `set_xlabel`, `set_ylabel`. Um bessere Abstände zwischen den Plots zu erreichen, können wir die Methode `tight_layout` auf einer oder mehrere `Axes-Objekte` anwenden. Die Methoden erwarten alle die optionalen Parameter `fontsize` und `fontweight`.

```python
plt.figure(figsize=(22, 5))
fig, ax  = plt.subplots(2, 1)


plt.show()
```

## Legenden in Subplots
Um Legenden in den Subplots anzuzeigen, können wir den Plot-Methoden wie gewohnt das Argument `label` mitgeben. Anzeigen können wir die Labels dann, in dem wir auf die `Axes-Objekte` die Methode `legend()` anwenden. Mit der Methode `grid` können wir wie gewohnt ein Gitter auf dem Plot anzeigen lassen.

`ax[1].scatter(x, y, label="a1 label")`         
`ax[1].legend()`

```python
fig, ax  = plt.subplots(2, 1)
ax[0].scatter(x,y, label="a0 label")
ax[1].scatter(x, y, label="a1 label")

ax[0].set_title("Axes Plot 0")
ax[1].set_title("Axes Plot 1")

ax[0].set_xlabel("X Label 0")
ax[1].set_xlabel("X Label 1")

ax[0].set_ylabel("y Label 0")
ax[1].set_ylabel("y Label 1")

figure_0 = ax[0].get_figure()
fig.tight_layout()

figure_1 = ax[1].get_figure()
fig.tight_layout()


ax[1].grid(axis='y')
plt.show()
```

## SuperTitle
man kann der `Figure` mit `fig.suptitle` noch einen Super-Titel geben, also eine Art Über-Schrift, die noch über den Axes-Titeln in der Hierachie steht.

```python
fig, ax  = plt.subplots(2, 1)
fig.suptitle("Zwei Plots", fontsize=14, fontweight="bold")

ax[0].scatter(x,y, label="a0 label")
ax[1].scatter(x, y, label="a1 label")

ax[0].set_title("Axes Plot 0",fontsize=11, fontweight="bold")
ax[1].set_title("Axes Plot 1",fontsize=11, fontweight="bold")

ax[0].set_xlabel("X Label 0", fontsize=9)
ax[1].set_xlabel("X Label 1", fontsize=9)

ax[0].set_ylabel("y Label 0")
ax[1].set_ylabel("y Label 1")

figure_0 = ax[0].get_figure()
fig.tight_layout()

figure_1 = ax[1].get_figure()
fig.tight_layout()

ax[0].legend()
ax[1].legend()


plt.show()
```
