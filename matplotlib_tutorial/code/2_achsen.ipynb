# Achsenbeschriftungen, Titel und Limits

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi * 2, 30)
y = np.sin(x)
plt.figure(figsize=(14,4))
plt.plot(x, y)
plt.show()
```

## Teilbereiche plotten

## X-Label, Y-Label
die Beschriftung für die X-Achse können wir mit `xlabel()` ändern, für die y-Achse entsprechend mit `ylabel()`.

## Legende
wir können jedem Plot ein Label mitgeben, welches dann in einer `Legend` anzeigt wird. In den `label`-elementen ist `Latex` erlaubt. Der Methode `legend` können wir optional noch das Argument `loc` mitgeben, welches den angezeigten Ort der Legende angibt. `upper right` steht für oben rechts, `lower left` wäre unten links. Wird für `legend` kein Argument angegeben, wählt Matplotlib die Position der Legende selbständig.

## Figur - Titel
