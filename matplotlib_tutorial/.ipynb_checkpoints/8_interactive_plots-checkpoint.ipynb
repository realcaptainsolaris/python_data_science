# Interactive Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/tips.csv")
df.head()
```

```python
# Matplotlib Scatter Plot
plt.scatter(x=df["total_bill"], y=df["tip"], marker="d")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
cov = [
    [2, 1],
    [1, 2]
]
mean = [1, 1]
dataset = np.random.multivariate_normal(mean=mean, cov=cov, size=100)
plt.scatter(x=dataset[:,0], y=dataset[:,1])
plt.show()
```
