# Diagramme mit Matplotlib

Wir betrachten folgende Diagramme:
- **Tortendiagramm (Pie Chart)**
- **Balkendiagramm (Bar Chart)**
- **Gestapeltes Balkendiagramm (Stacked Bar Chart)**


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Tortendiagramm
Das Snippet erstellt ein Tortendiagramm, das die prozentuale Verteilung der Früchte Apfel, Banane, Kirsche und Dattel anhand der Werte in sizes darstellt, wobei jede Kategorie eine spezifische Farbe (colors) erhält und die Prozente (autopct='%1.1f%%') direkt auf die Segmente geschrieben werden, während startangle=140 den Startwinkel des Diagramms festlegt.

```python
# Tortendiagramm (Pie Chart)

labels = ['Apfel', 'Banane', 'Kirsche', 'Dattel']
sizes = [30, 25, 35, 10]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%0.2f%%', startangle=140)
plt.title("Fruchtverteilung")
plt.show()
```

## Balkendiagramm

```python
# Balkendiagramm (Bar Chart)

categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
colors = ['red', 'green', 'blue', 'purple']

plt.figure(figsize=(8,5))
plt.bar(categories, values, color=colors, edgecolor='black', width=1)  # default ist 0.8

plt.xlabel("Kategorien")
plt.ylabel("Werte")
plt.title("Einfaches Balkendiagramm")
plt.show()
```

## Horizontales Balkendiagramm (Bar Chart)

```python
# horizontales Balkendiagramm (Bar Chart)

categories = np.array(['A', 'B', 'C', 'D'])
values = np.array([4, 7, 1, 8])

sorted_idx = np.argsort(values) # hier einsatz von argsort zeigen

plt.figure(figsize=(8,5))
plt.barh(categories[sorted_idx], values[sorted_idx], color='skyblue', edgecolor='black',  height=1)  # default ist 0.8
plt.ylabel("Kategorien")
plt.xlabel("Werte")
plt.title("Einfaches horizontales Balkendiagramm")
plt.show()
```

## Gestapeltes Balkendiagramm (Stacked Bar Chart)

```python
# Gestapeltes Balkendiagramm (Stacked Bar Chart)

labels = ['Gruppe 1', 'Gruppe 2', 'Gruppe 3']
men_means = [20, 35, 30]
women_means = [25, 32, 34]

x = np.arange(len(labels))

plt.figure(figsize=(8,5))
plt.bar(x, men_means, label='Männer', color='blue')
plt.bar(x, women_means, bottom=men_means, label='Frauen', color='pink')

plt.xticks(x, labels)
plt.xlabel("Gruppen")
plt.ylabel("Werte")
plt.title("Gestapeltes Balkendiagramm")
plt.legend()
plt.show()
```
