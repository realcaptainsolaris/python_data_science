# Achsenbeschriftungen, Titel und Limits

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi * 2, 30)
y = np.sin(x)
plt.figure(figsize=(14,4))
plt.plot(x, y)
plt.show()
```

## X-Ticks / Y-Ticks festlegen
im Beispiel setzen wir die Werte für die X- und Y-Achsenbezeichner selber und vergrößern damit den Canvas.

```python
plt.figure(figsize=(14,4))
plt.plot(x, y)
plt.xticks(range(-2, 10))
plt.yticks(range(-2, 3))
plt.show()
```

## Darstellungen gleichgewichten (Equal axis aspect ratio)

```python
# Kreis definieren
theta = np.linspace(0, 2*np.pi, 300)
x = np.cos(theta)
y = np.sin(theta)

# Subplots erstellen
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Erster Plot ohne gleichgewichtete Achsen
axes[0].plot(x, y, label="Kreis")
axes[0].set_title("Ohne gleichgewichtete Achsen")

# Zweiter Plot mit gleichgewichteten Achsen
axes[1].plot(x, y, label="Kreis")
axes[1].axis("equal")  # Gleichgewichtete Achsen setzen
axes[1].set_title("Mit gleichgewichteten Achsen")
```

## Teilbereiche plotten
Wir können den zu Plottenden Bereich mit `xlim` und `ylim` einschränken.

```python
plt.figure(figsize=(14,4))
plt.plot(x, y)
plt.xlim(0, 2)
plt.ylim(0, 1)
plt.show()
```

## X-Label, Y-Label
die Beschriftung für die X-Achse können wir mit `xlabel()` ändern, für die y-Achse entsprechend mit `ylabel()`.

```python
plt.plot(x, y)
plt.xlabel("x-Values auf der X-Achse")
plt.ylabel("y-Values auf der Y-Achse")
plt.axis("equal")
plt.show()
```

## Legende
wir können jedem Plot ein Label mitgeben, welches dann in einer `Legend` anzeigt wird. In den `label`-elementen ist `Latex` erlaubt. Der Methode `legend` können wir optional noch das Argument `loc` mitgeben, welches den angezeigten Ort der Legende angibt. `upper right` steht für oben rechts, `lower left` wäre unten links. Wird für `legend` kein Argument angegeben, wählt Matplotlib die Position der Legende selbständig.

```python
plt.figure(figsize=(14,4))
plt.plot(x, np.sin(x), label="$sin(x)$", linestyle="dashed")
plt.plot(x, np.sin(x**2), label="$sin(x^2)$")
plt.legend(loc="lower right")
plt.show()
```

## Figur - Titel
Wir können der Figur einen Titel geben. In unserem einfachen Fall hier, ist das auch gleichzeitig der Titel für die Axes. Später werden wir sehen, dass wir auch den einzelnen Axes ebenso Titel zuweisen können. Mit `fontsize` können wir den Schriftgrad bestimmen.

```python
plt.figure(figsize=(14,4))
plt.title("Beispielplot mit sinus-Funktion", fontsize=20)
plt.plot(x, np.sin(x), label="$sin(x)$")
plt.plot(x, np.sin(x**2), label="$sin(x^2)$")
plt.xlabel("x-Values")
plt.ylabel("y-Values")
plt.legend()
plt.show()
```
