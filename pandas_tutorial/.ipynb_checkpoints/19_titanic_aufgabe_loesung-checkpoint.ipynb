# Beispiel Aufgabe: Titanic Dataset, Lösung

es sollen Aufgaben zum Titanic Dataset ausgeführt werden.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

df = pd.read_csv("data/titanic.csv", usecols=['PassengerId', 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked'])
df.set_index("PassengerId", inplace=True)
df
```

## Aufgabe 1
Ersetze fehlende Age-Werte mit -1    
Ersetze fehlende Pclass-Werte mit -1   
Ersetze fehlende Cabin-Werte mit "None"
setze Datentyp für Spalte Age und Pclass auf int8   

```python
# mit boolscher Serire
missing_pclass = df["Pclass"].isna()
df.loc[missing_pclass, 'Pclass'] = -1

# via fillna inplace
df.Age = df.Age.fillna(-1)
df.Cabin = df.Cabin.fillna("None")

df[["Age", "Pclass"]] = df[["Age", "Pclass"]].astype("int8")
df.head(10)
```

## Aufgabe 2: 
Lösche alle restlichen Zeilen mit `nan-Werten`.

```python
df.dropna(inplace=True)
df.isna().any()
```

## Aufgabe 3
Berechne für folgende Gruppierungen die min, max, median
* Survived und männlich
* Not Survived und männlich
* Survived und weiblich
* Not Survived und weiblich

```python
# via Filtern (umständlich, hier im Beispiel nur für Male und Survived)
df_survived_and_male  = df[(df.Sex == "male") & (df.Survived == True)]
df_survived_and_male["Age"].describe()
```

```python
# Gruppieren und Aggregieren
result = df.groupby(["Sex", "Survived"]).agg({'Age': ['median', 'min', 'max'], 'Survived': ['sum']})
result
```

## Aufgabe 3
Angehörige welchen Alters haben am häufigsten überlebt?

```python
filter_survived = df[(df.Survived == True) & (df.Age > 0)]
survived_ages = filter_survived.Age.value_counts().nlargest(1)
survived_ages, survived_ages.values[0]
```
